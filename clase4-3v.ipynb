Es una manera más rápida de tratar con gran cantidad de datos. La manera en la que se agrupan los datos es por *clases*.

Se suele usar entre 5 y 20 clases y se sugiere que todas tengan el mismo tamaño.

Los valores que limitan a cada una de las clases se conocen como *límites de clases*.

El *ancho de clase* es la diferencia entre los límites de cada clase:

$$ \text{ancho de clase}=\text{límite superior}-\text{límite inferior} $$

$$ \text{límite superior}=\text{ancho de clase}+\text{límite inferior} $$

Estamos tomando el ancho de la clase como 0.1 m. Entonces los límites de clase son:

$$1.5,\ 1.6,\ 1.7,\ 1.8,\ 1.9,\ 2.0$$

Un dato pertenece a una clase si su valor numérico es estrictamente mayor que el límite inferior y menor o igual que el límite superior.

La *marca de clase* se obtiene dividiendo como el promedio de los límites de cada clase y es un valor representativo de los datos de esa clase:

$$ m_i=\dfrac{\text{límite superior}+\text{límite inferior}}{2} $$

| Número de clase | Intervalos de clase | $m_i$ | $f_i$ |
|-----------------|---------------------|-------|-------|

La **frecuencia absoluta** de clase es el total de los datos que pertenecen a cada clase.

La **frecuencia relativa** indica la proporción de los datos que pertenecen a cada clase.

La **frecuencia acumulada** es la cantidad de datos que indica cuántos de ellos tienen un valor numérico menor o igual que el límite superior hasta una clase determinada.

### Media aritmética, $\bar{x}$

$$\bar{x}=\dfrac{\sum_{i=1}^km_if_i}{\sum_{i=1}^kf_i}$$

donde $m_i$ son las marcas de clase, $f_i$ la frecuencia absoluta, y $k$ es el número de clases.

| Número de clase | Intervalos de clase | $m_i$ | $f_i$ | $ \left.m_i\cdot f_i\right.$ |
|-----------------|---------------------|-------|-------|-------------------------------|
| 1               | $[1.50,1.60)$       | 1.55  | 8     | 12.40                         |

La suma de columna $m_i\cdot f_i$ es 52.5

Lo dividimos entre 30 para obtener la media aritmética.

donde $m_i$ son las marcas de clase, $\bar{x}$ la media aritmética, $f_i$ la frecuencia absoluta, $k$ es el número de clases, y $n$ es el número total de datos.

### Histograma

Es una gráfica de una variable estadística en forma de barras sin espaciamiento entre ellas, donde la altura de la barra es la frecuencia (absolutas o relativas) y el ancho de cada barra representa los límites de clase.
